```python
from IPython.display import Image, display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from time import sleep
from PIL import Image as Imag
import pyautogui
from math import ceil
```

```python
def perform(box):
    '''Moves the mouse to the box location.
    Performs one tile mousedrag in for directions '''
    pyautogui.moveTo(box[0], box[1])
    pyautogui.drag(-60, 0, 0, button='right')
    pyautogui.move(60, 0)
    pyautogui.drag(60, 0, 0, button='right')
    pyautogui.move(-60, 0)
    pyautogui.drag(0, 60, 0, button='right')
    pyautogui.move(0, -60)
    pyautogui.drag(0, -60, 0, button='right')
    pyautogui.move(0, 60)
```
