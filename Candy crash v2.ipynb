```python
from IPython.display import Image, display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from time import sleep
from PIL import Image as Imag
import pyautogui
```

```python
def perform():
    pyautogui.drag(-60, 0, 0, button='right')
    pyautogui.move(60, 0)
    pyautogui.drag(60, 0, 0, button='right')
    pyautogui.move(-60, 0)
    pyautogui.drag(0, 60, 0, button='right')
    pyautogui.move(0, -60)
    pyautogui.drag(0, -60, 0, button='right')
    pyautogui.move(0, 60)
```

really 500x524

```python
tile_w = 500/7
tile_h = 524/8
```

```python
img ='Images/btn_close.png'
display(Image(img))
```

```python
offset_x = tile_w/2
offset_y = tile_h/2
```

```python
board_w = int(input('Tiles width: '))
```

```python
board_h = int(input('Tiles height: '))
```

```python
while True:
    if pyautogui.locateOnScreen('Images/map.png', confidence=0.7):
        break
    if pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7):
        x, y = pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7)
        pyautogui.click(x/2, y/2, duration=0.3, button='right')
        break
    start_x = tile_w+offset_x
    start_y = pyautogui.size().height - tile_h * (board_h-1) + offset_y
    pyautogui.moveTo(start_x, start_y) 
    for i in range(ceil(board_h/2)):
        if pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7):
            x, y = pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7)
            pyautogui.click(x/2, y/2, button='right')
            break
        for j in range(ceil(board_w/2)):
            perform()
            pyautogui.move(tile_w*2, 0)
        start_y += tile_h*2   
        pyautogui.moveTo(start_x, start_y)
```
