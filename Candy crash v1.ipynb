```python
from IPython.display import Image, display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from time import sleep
from PIL import Image as Imag
import pyautogui
```

```python
def perform():
    '''Perform four swipes aling x and y axis.'''
    pyautogui.drag(-60, 0, 0.1, button='right')
    pyautogui.move(60, 0)
    pyautogui.click()
    pyautogui.drag(60, 0, 0, button='right')
    pyautogui.move(-60, 0)
    pyautogui.drag(0, 60, 0, button='right')
    pyautogui.move(0, -60)
    pyautogui.drag(0, -60, 0, button='right')
    pyautogui.move(0, 60)
```

```python
def closeButtonLookup():
    if bool(pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7))==True:
        x, y = pyautogui.locateCenterOnScreen('Images/btn_close.png', confidence=0.7)
        pyautogui.click(x/2, y/2, duration=0.5, button='right')
        print('OK,I am out of here!')
        return True
    return False
```

```python
tile_w = 500/7
tile_h = 524/8
```

```python
offset_x = tile_w/2
offset_y = tile_h/2
```

```python
board_w = int(input('Tiles width: '))
```

```python
board_h = int(input('Tiles height: '))
```

```python
while True:
#     See if we are on the map yet
    if bool(pyautogui.locateOnScreen('Images/map.png', confidence=0.7))==True:
        print("I'm back to the map.")
        break
    if closeButtonLookup()==True:
        break
#     Set the start points
    start_x = tile_w+offset_x
    start_y = pyautogui.size().height - tile_h * (board_h) + offset_y
    pyautogui.moveTo(start_x, start_y) 
    for i in range((board_h)):
        if closeButtonLookup()==True:
            break
#         Step Horisontally 
        for j in range(board_w-2):
            perform()
            pyautogui.move(tile_w, 0)
#         Step vertically one tile lower
        start_y += tile_h  
        pyautogui.moveTo(start_x, start_y)
```
